```python
import pandas as pd
```

```python
df=pd.read_csv("/content/weather_forecast_data.csv")
df
```

```python
df['Rain']=df['Rain'].replace({'rain':0,'no rain':1})
```

```python
df
```

```python
y=df['Rain']
y
```

```python
x=df.drop('Rain',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = Perceptron(max_iter=1000, eta0=1.0, random_state=42)
model.fit(x_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
cr
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_p=model.predict(x_test)
y_p
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_p)
accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
y_pd=model.predict(x_test)
y_pd
```

```python
from sklearn.utils import resample
l_acc=[]
s_acc=[]
k_acc=[]
for i in range(0,20):
  x_r,y_r=resample(x,y,n_samples=200)
  x_tr,x_te,y_tr,y_te=train_test_split(x_r,y_r,test_size=0.2,random_state=42)
  model=LogisticRegression()
  model.fit(x_tr,y_tr)
  y_pred=model.predict(x_te)
  accuracy=accuracy_score(y_te,y_pred)
  l_acc.append(accuracy)
  svc=SVC()
  svc.fit(x_tr,y_tr)
  y_p=svc.predict(x_te)
  accuracy=accuracy_score(y_te,y_p)
  s_acc.append(accuracy)
  knn=KNeighborsClassifier(n_neighbors=3)
  knn.fit(x_tr,y_tr)
  y_pr=knn.predict(x_te)
  accuracy=accuracy_score(y_te,y_pr)
  k_acc.append(accuracy)
```

```python
import matplotlib.pyplot as plt
plt.plot(k_acc,label="knn")
plt.plot(s_acc,label="svc")
plt.plot(l_acc,label="log")
plt.legend()
```
