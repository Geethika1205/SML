```python
df=pd.read_csv("/content/loan_data.csv")
```

```python
df['person_gender']=df['person_gender'].replace({'male':0,'female':1})
df['person_education']=df['person_education'].replace({'High School':0,'Bachelor':1,'Master':2,'University':3,'Associate':4})
df['person_home_ownership']=df['person_home_ownership'].replace({'RENT':0,'OWN':1,'MORTGAGE':2})
df['loan_intent']=df['loan_intent'].replace({'PERSONAL':0,'EDUCATION':1,'MEDICAL':2,'HOMEIMPROVEMENT':3,'DEBTCONSOLIDATION':4,})
df['previous_loan_defaults_on_file']=df['previous_loan_defaults_on_file'].replace({'No':0,'Yes':1})

```
